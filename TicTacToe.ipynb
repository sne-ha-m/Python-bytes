```python
#Tic tac Toe Game
print("Let's play Tic Tac Toe")
print('You want to start? Y or N')
ans = input()

if ans.lower() == 'y':
    print('Let us start')
    from IPython.display import clear_output

    def display_board(board):
        clear_output()
    
        print(' ' + board[6] + ' | ' + board[7] + ' | ' + board[8])
        print('-----------')
        print(' ' + board[3] + ' | ' + board[4] + ' | ' + board[5])
        print('-----------')
        print(' ' + board[0] + ' | ' + board[1] + ' | ' + board[2])
    board = ['1','2','3','4','5','6','7','8','9']
    
    def player_input():
        global n1
        global n2
        n1 = input('Enter yor name for Player 1:\t') 
        n2 = input('Enter yor name for Player 2:\t') 
        op = input(f'{n1}: You want to be X or O?').lower()
        if op == 'x' or op == 'o':
            print('Let us begin')
            if op == 'x':
                return ('X', 'O')
            else:
                return ('O', 'X')
        else:
            print('Invalid input. pls try again')    
            player_input()
            
    def place_marker(board,option,position):
        board[position] = option
        
    def space_check(board, position):
        if board[position] == 'X' or board[position] == 'O'  :
            return False
        else:
            return True
    
    def full_board_check(board):
        for i in range(0,9):
            if space_check(board, i):
                return False
        return True
        
    def win_check(board,mark):
        return ((board[6] == mark and board[7] == mark and board[8] == mark) or # across the top
                (board[3] == mark and board[4] == mark and board[5] == mark) or # across the middle
                (board[0] == mark and board[1] == mark and board[2] == mark) or # across the bottom
                (board[6] == mark and board[3] == mark and board[0] == mark) or # down the middle
                (board[7] == mark and board[4] == mark and board[1] == mark) or # down the middle
                (board[8] == mark and board[5] == mark and board[2] == mark) or # down the right side
                (board[6] == mark and board[4] == mark and board[2] == mark) or # diagonal
                (board[8] == mark and board[4] == mark and board[0] == mark)) # diagonal
    
    op = player_input()
    P1,P2 = op
    print('The sign of {} is {} and {} is {}'.format(n1,P1,n2,P2))
    start = input(f'Who wants to play first? {n1} or {n2}').lower()
    game_on = 'on'
    
    while game_on == 'on':
        
        if start == n1:
    
            display_board(board)
            position = int(input('Choose from 1-9 where you want to position your mark as displayed in the positing board'))
            place_marker(board,op[0], position-1)
            if win_check(board,op[0]):
                display_board(board)
                print(f"Congratulations!! {n1} wins")
                game_on = 'off'
                break
                
            else:
                
                if full_board_check(board):
                    display_board(board)
                    print('The game is a draw!')
                    break
                    
                else:
                    start = n2
                #print('Player 2 Turn\n')
            
            
        else:
            display_board(board)
            position = int(input('Choose from 1-9 where you want to position your mark as displayed in the positing board'))
            place_marker(board,op[1], position-1)
            if win_check(board,op[1]):
                display_board(board)
                print(f"Congratulations!! {n2} wins")
                game_on = 'off'
                break
    
            else:
            
                if full_board_check(board):
                    display_board(board)
                    print('The game is a draw!')
                    break
                    
                else:
                    start = n1
                #print('Player 1 Turn\n')
            
            
    

       
    
else:
    print('See you again')
```
