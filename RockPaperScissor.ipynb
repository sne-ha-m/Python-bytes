```python
#rock paper scissors game
import random
from IPython.display import clear_output

#my_win = 0
#pc_win = 0

my_game = 3

my_t_score = 0
pc_t_score = 0

my_total_score = 0
pc_total_score = 0



def game():
    
    my_score = 0
    pc_score = 0
    
    while my_score < 2  and pc_score < 2:
        
        
        my_choice = input('Choose one from Rock/ Paper / Scissor').lower()
        pc_choice = random.choice(['rock','paper','scissor'])
        
        #clear_output()
        
        if my_choice == pc_choice:
            print('Draw')
            #print(f'score: {my_score} and pc score: {pc_score}')
            print(f'You guessed {my_choice} and pc guessed: {pc_choice}')
            
        
        else:
        
            if my_choice == 'rock' and pc_choice == 'scissor':
                my_score += 1
                #print(f'score: {my_score} and pc score: {pc_score}')
                print(f'You guessed {my_choice} and pc guessed: {pc_choice}')
            
            elif my_choice == 'scissor' and pc_choice == 'paper':
                my_score += 1
                #print(f'score: {my_score} and pc score: {pc_score}')
                print(f'You guessed {my_choice} and pc guessed: {pc_choice}')
                
            elif my_choice == 'paper' and pc_choice == 'rock':
                my_score += 1
                #print(f'score: {my_score} and pc score: {pc_score}')
                print(f'You guessed {my_choice} and pc guessed: {pc_choice}')
            
            elif my_choice not in ['rock','paper','scissor']:
                print('Wrongly Spelt. Check again')
                
            else:
                pc_score += 1
                #print(f'score: {my_score} and pc score: {pc_score}')
                print(f'You guessed {my_choice} and pc guessed: {pc_choice}')
    
    
    print(f'Your score is {my_score} and PC score is {pc_score}') 
    
    return [my_score, pc_score]


    
def tournament():
    my_tournament_win = 0
    pc_tournament_win = 0
    score = game()
    clear_output()
    
    if score[0] > score[1]:
        my_tournament_win += 1
    else:
        pc_tournament_win += 1
    
    #my_list = [my_tournament_win, pc_tournament_win]  
    print(f'this tournament scores: {my_tournament_win} and {pc_tournament_win}')
    my_score = score[0] 
    pc_score = score[1]
    #print( [my_tournament_win] + [pc_tournament_win] +[my_score]+[pc_score])
    return [my_tournament_win] + [pc_tournament_win] +[my_score]+[pc_score]
        


while my_game > 0:

    x = tournament()
    my_t_score += int(x[0])
    pc_t_score += int(x[1])
    my_total_score += int(x[2])
    pc_total_score += int(x[3])
    print(f'Yor total win till now: {my_t_score} and pc total win till now: {pc_t_score}')
    print(f'Your total score till now: {my_total_score} and pc total score till now: {pc_total_score}')
    my_game -= 1

    clear_output()
print(f'..............tournament scores...............)
if my_t_score > pc_t_score:
    
    print('Congrats! you won the tournament!!')
    print(f'Your total score for all tournaments: {my_total_score} and total tournament scores: {my_t_score}')
else:
    print('PC won!')
    print(f'PC total score for all tournaments: {pc_total_score} and total tournament score: {pc_t_score}')
```
